```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
array = df.values
df.columns
```

```python
for i in range(len(df['Evening Available'])):
    if df['Evening Available'][i] != 'Yes' and df['Evening Available'][i] !='No':
        df.loc[i,('Evening Available')]='Yes'
        break
def mapper(label):
    mapping = {l:idx for idx,l in enumerate(np.unique(df[label]))}
    df[label] = df[label].map(mapping)
df.head()
```

```python
mapper('Occupation')
mapper('Type')
mapper('Gender')
mapper('Morning Available')
mapper('Location')
mapper('Afternoon Available')
mapper('Evening Available')
mapper('Qualified')
df.fillna(0,inplace=True)
df.loc[df['Doorstep Service']!=0,'Doorstep Service']=1
#mapper('Doorstep Service')
```

```python
userIDs = []
def userRating(labelname):
    l = []
    for i in user_df[labelname]:
        #l = list(map(int,user_df[labelname]))
        if user_df[labelname].dtype == 'int64':
            pass
        else:
            l.append(int(i.replace(',','')))
            
    return l

a = set(userRating('CustomerID1'))
b = set(userRating('CustomerID2'))
c = set(userRating('CustomerID3'))
d = set(userRating('CustomerID4'))
e = set(userRating('CustomerID5'))

userIDs.extend(list(a))
userIDs.extend(list(b))
userIDs.extend(list(c))
userIDs.extend(list(d))
userIDs.extend(list(e))
```
