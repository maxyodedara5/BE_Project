```python
# import all the required packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
from sklearn.metrics.pairwise import cosine_distances
```

```python
# reading data from dataset
df = pd.read_csv('data.csv')
# breif about the columns in the dataset
df.columns
```

```python
# Removing outliers and making Series consistent
for i in range(len(df['Evening Available'])):
    if df['Evening Available'][i] != 'Yes' and df['Evening Available'][i] !='No':
        df.loc[i,('Evening Available')]='Yes'
        break
# Mapping Strings etc to classes
def mapper(label):
    mapping = {l:idx for idx,l in enumerate(np.unique(df[label]))}
    df[label] = df[label].map(mapping)
print(df.head(2))

# Removing NaN and filling data points with '0'
df.fillna(0, inplace=True)
```

```python
# Mapping all the Series
mapper('Occupation')
mapper('Type')
mapper('Gender')
mapper('Morning Available')
mapper('Location')
mapper('Afternoon Available')
mapper('Evening Available')
mapper('Qualified')
```

```python
df.head()
```

```python
# Droping all unnecessary Series
df = df.drop(['Phone number', 'Serial number','Aadhar Card'],axis=1)
```

```python
# Grouping w.r.t locations
"""As to all workers available in a particular location area"""
occupations=defaultdict(list)

def all_occupations_in_a_location():
    for index,row in df.iterrows():
        occupations[row['Location']].append(row['Occupation'])
        occupations[row['Location']]
    for key, values in occupations.items():
        t_set = set(occupations.get(key))
        occupations[key] = list(t_set)
all_occupations_in_a_location()
print(occupations.keys())
df.head()
```
