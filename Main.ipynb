```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
array = df.values
df.columns
```

```python
for i in range(len(df['Evening Available'])):
    if df['Evening Available'][i] != 'Yes' and df['Evening Available'][i] !='No':
        df.loc[i,('Evening Available')]='Yes'
        break
def mapper(label):
    mapping = {l:idx for idx,l in enumerate(np.unique(df[label]))}
    df[label] = df[label].map(mapping)
df.head()
```

```python
mapper('Occupation')
mapper('Type')
mapper('Gender')
mapper('Morning Available')
mapper('Location')
mapper('Afternoon Available')
mapper('Evening Available')
mapper('Qualified')
```
